```python
#Import dos gráficos
import matplotlib
from matplotlib import transforms, pyplot as plt
import numpy as np

%matplotlib inline
#%matplotlib
```

```python
# Definindo cores
GRAY1, GRAY2, GRAY3 = '#231F20', '#414040', '#555655'
GRAY4, GRAY5, GRAY6 = '#646369', '#76787B', '#828282'
GRAY7, GRAY8, GRAY9 = '#929497', '#A6A6A5', '#BFBEBE'
BLUE1, BLUE2, BLUE3, BLUE4 = '#174A7E', '#4A81BF', '#94B2D7', '#94AFC5'
RED1, RED2 = '#C3514E', '#E6BAB7'
GREEN1, GREEN2 = '#0C8040', '#9ABB59'
ORANGE1 = '#F79747'
```

```python
# Fontes para Arial
plt.rcParams['font.family'] = 'Arial'
matplotlib.rcParams['mathtext.fontset'] = 'custom'
matplotlib.rcParams['mathtext.bf'] = 'Arial:bold'
matplotlib.rcParams['mathtext.it'] = 'Arial:italic'
```

```python
# Séries de Zueira
X = ['Jan', 'Fev', 'Mar', 'Abr', 'Mai', 'Jun',
     'Jul', 'Ago', 'Set', 'Out', 'Nov', 'Dez']
Y = [80,70,55,30,30,20,80,70,60,49,45,55]
Z = [20,30,45,60,70,80,20,30,40,51,55,45]
```

```python
# Criando uma figura nova
fig, ax1 = plt.subplots(figsize=(8.2, 6.09), dpi=110)             

# Ajustando o subplot
fig.subplots_adjust(left=0.104, right=0.768, top=0.751, bottom=0.187)

# Linha do Ticket 1
ax1.plot(range(12), Z, linewidth=4, color=BLUE1)
ax1.scatter(range(12)[-5:], Z[-5:], s=90, color=BLUE1, clip_on=False)

# Desenhando a linha do ticket 2 
ax1.plot(range(12), Y, linewidth=4, color=RED1)
ax1.scatter(range(12)[-5:], Y[-5:], s=90, color=RED1, clip_on=False)

# Processado
for i, v in enumerate(Z[-5:]):
    ax1.annotate(str(v),
                 (i+7, v),                    # (x,y) point to annotate
                 xytext=(12,-5),            # (x,y) to place the text at
                 textcoords='offset points',  # offset (in points)
                 color=BLUE2,
                 fontsize=15)

# Recebido
for i, v in enumerate(Y[-5:]):
    ax1.annotate(str(v),
                 (i+7, v),                   # (x,y) point to annotate
                 xytext=(12, -5),          # (x,y) to place the text at
                 textcoords='offset points', # offset (in points)
                 color=RED1,
                 fontsize=15)


# Limites dos eixos
ax1.set_xlim([0, 11])
ax1.set_ylim([0, 100])

# Propriedades dos eixos
plt.setp(ax1,
         xticks=list(range(12)),
         xticklabels=X,
         yticks=list(range(0, 110, 10)))

# X e Y tick aparência
for i in ax1.get_yticklabels() + ax1.get_xticklabels():
    i.set_fontsize(14)
    i.set_color(GRAY3)

# Movendo eixo y
for item in ax1.get_yticklabels():
    offset = transforms.ScaledTranslation(-0.07, 0, fig.dpi_scale_trans)
    item.set_transform(item.get_transform() + offset)

# Mudando a aparência no geral
ax1.tick_params(bottom='off')

# Trocando cores de spines
ax1.tick_params(color=GRAY4)
ax1.spines['bottom'].set_color(GRAY3)
ax1.spines['left'].set_color(GRAY3)

# Removendo bordas
ax1.spines['right'].set_visible(False)
ax1.spines['top'].set_visible(False)


# Linha de referência vertical
plt.axvline(x=6, ymax=1, color='#636369', linewidth=1)

'''
# Legendando diretamente
ax1.text(11.5, 172, 'Received', fontsize='21', color=GRAY6)
ax1.text(11.5, 135, 'Processed', fontsize='21', color=BLUE1)
'''


# Título do eixo
#ax1.text(-1.6, 325, 'Ticket volume over time', fontsize=19, color=GRAY4)
ax1.text(-1.55, 30, 'Intenção de votos (%)', fontsize=14, rotation='vertical', color=GRAY4)
ax1.text(-0.34, -12, '2021', fontsize=11, color=GRAY4)


# Título do gráfico
ax1.text(-1.6, 120, '$\\bf{Dilma \ ganhará \ segundo \ turno \ contra \ Bolsonaro}$',
         fontsize=25, color=GRAY4)
ax1.text(-1.6, 110, 'aponta Instituto DataGuedes.',
         fontsize=16, color=GRAY7)

# Texto informativo
'''
ax1.text(4, 252,
         '$\\bf{2\ employees\ quit\ in\ May}$. We nearly kept up with'
         ' incoming\n'
         'volume in the following two months, but fell behind with the\n'
         'increase in Aug and haven''t been able to catch up since.',
         fontsize=14, color=GRAY5, linespacing=1.4)
'''

ax1.text(8, 90,
         'Depois que $\\bf{Guilherme \ Julio}$ colocou \n'
         'seu incondicionável apoio a ex-presidente, \n'
         'o jogo virou em favor dela. A previsão é que \n'
         'isso não permaneça nos próximos meses.',
         fontsize=11.5, color=GRAY3, linespacing=1.3)

ax1.text(9, 70,
         'Julio ficou impressionado com o projeto \n'
         'de estocar o vento, em meio a $\\bf{pior \ crise}$ \n'
         '$\\bf{energética \ do \ país.}$',
         fontsize=11.5, color=GRAY3, linespacing=1)

ax1.text(5, 105,
         'Apoio de Julio',
         fontsize=11.5, color=GRAY3, linespacing=1.3)

# Fonte
ax1.text(-1.6, -30,
         'Cerca de 1500 pessoas foram entrevistadas \n'
         'entre 12:50 - 12:56 do dia 14/07/2021. \n'
         'A confiabilidade dos dados é de 95%.',
         fontsize=11.5, color=GRAY3, linespacing=1.3)
```
